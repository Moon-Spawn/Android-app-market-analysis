```python
import pandas as pd
apps = pd.read_csv("C:\\Users\\parth\\Projects\\Android app market analysis\\data\\googleplaystore.csv")

#Removing Duplicates
apps = apps.drop_duplicates()

display(apps.head(5))
print('the total number of apps are: ', len(apps))
print('\n')
print( apps.info())    
```

```python
#Droping Null values from columns which have few
apps.dropna(subset=['Type', 'Content Rating', 'Current Ver', 'Android Ver'], inplace = True)

import numpy as np

#Treating uncertain size as missing value
apps = apps.replace({'Varies with device': np.nan})

apps.info()
```

```python
# List of characters to remove
chars_to_remove = [',', '$', '+']
# List of column names to clean
cols_to_clean = ['Installs', 'Price']

for col in cols_to_clean:
    for char in chars_to_remove:
        apps[col] = apps[col].astype(str).str.replace(char, '')
    # Convert col to numeric
    apps[col] = pd.to_numeric(apps[col])

#Converting Size to numeric by replacing 'k' and 'M' by appropriate values    
apps['Size'] = (apps.Size.replace(r'[kM]+$', '', regex = True).astype(float) *\
             apps.Size.str.extract(r'[\d\.]+([kM]+)', expand = False).fillna(1).replace(['k', 'M'], [10**3, 10**6]).astype(int))
    
apps.info()
```

```python
import plotly
plotly.offline.init_notebook_mode(connected = True)
import plotly.graph_objs as go

#Number of categories
print("Number of Categories: ", len(apps['Category'].unique()))

num_apps_cat = apps['Category'].value_counts().sort_values(ascending = False)

data = [go.Bar(x = num_apps_cat.index, y = num_apps_cat.values)]
layout = {'title':'Number of apps by categories', 'xaxis': {'title': 'Categories'},
         'yaxis':{'title': 'Number of apps'}}

plotly.offline.iplot({'data' :data, 'layout':layout})
```

```python
print("average rating of apps: ", apps.Rating.mean())

data = [go.Histogram(x = apps['Rating'])]


layout = {'shapes': [{'type': 'line',
                     'x0': apps.Rating.mean(),
                     'y0': 0,
                     'x1': apps.Rating.mean(),
                     'y1': 1000,
                     'line': {'dash': 'dashdot'}
                     }],
          'title': 'Number of apps by Rating',
          'xaxis': {'title': 'Rating'},
          'yaxis': {'title': 'Numer of Apps'}
         }
plotly.offline.iplot({'data': data, 'layout':layout})
```

```python
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')

#Filter out rows with missing size or rating
apps_size_rating = apps[(~apps['Size'].isnull())& (~apps['Rating'].isnull())]

#Subset having large categories only
large_cats=apps_size_rating.groupby('Category').filter(lambda x: len(x)>=250).reset_index()

plt1 = sns.jointplot(x = large_cats['Size'], y = large_cats['Rating'], kind = 'hex')

paid_apps = large_cats[large_cats['Type'] == 'Paid']

plt2 = sns.jointplot(x = large_cats['Price'], y = large_cats['Rating'])
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
fig.set_size_inches(15,8)

#dataframe containing only popular categories
pop_apps = apps[apps.Category.isin(['GAME', 'FAMILY', 'PHOTOGRAPHY', 'MEDICAL', 'TOOLS', 'FINANCE', 'LIFESTYLE','BUSINESS'])]

ax = sns.stripplot(x = pop_apps['Price'], y = pop_apps['Category'], jitter =True, linewidth = 1)

ax.set_title('App pricing by category')
```

```python
#filtering out junk apps from popular apps dataframe
pop_apps2 = pop_apps[pop_apps.Price < 100]

fig, ax= plt.subplots()
fig.set_size_inches(15,8)

ax = sns.stripplot(x = pop_apps2['Price'], y = pop_apps2['Category'], jitter =True, linewidth = 1)

ax.set_title('App pricing by category')
```

```python
#dataframe for installs of paid apps
paid = go.Box(y = apps[apps['Type']=='Paid']['Installs'], name = 'Paid')
#dataframe for installs of free apps
free = go.Box(y = apps[apps['Type']=='Free']['Installs'], name = 'Free')
layout = go.Layout(title = 'Number of installs paid vs free apps', 
                   yaxis = dict(title = 'Number of installs', 
                                type = 'log', autorange = True)
                  )
data = [paid, free]

plotly.offline.iplot({'data':data, 'layout':layout})
```
